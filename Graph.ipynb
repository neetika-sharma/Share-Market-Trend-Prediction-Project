```python
import numpy as np
import matplotlib.pyplot as plt
N = 5
gram1 = (82.27513227513228, 82.01058201058201, 84.6560846561, 82.27513227513228, 85.18518518518519)
gram2 = (85.71428571428571, 0, 85.1288359788, 84.65608465608465, 85.18518518518519)
gram3= (85.18518518518519, 0, 85.1851851852, 84.65608465608465, 82.53968253968254)


ind = np.arange(N) 
width = 0.25       
plt.bar(ind, gram1, width, label='1-Gram')
plt.bar(ind + width, gram2, width,label='2-Gram')
plt.bar(ind + width*2, gram3, width,label='3-Gram')


plt.title('Comparison of models')
plt.ylim(81, 88)
plt.xticks(ind + width/2 , ('LogReg', 'NaiveB', 'RanForest', 'Lin SVM', 'Non-lin SVM'))
plt.legend(loc='upper right')
plt.show()
```
